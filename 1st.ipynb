```python
import matplotlib.pyplot as plt 
import numpy as np

num_lines = 50 ; num_turns = 10 ; num_points = 1000

fig ,ax = plt.subplots(figsize=(6,6))
theta= np.linspace(0 , num_turns * 2 * np.pi , num_points)
r = np.linspace(0,1,num_points)

x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot(x , y , color = 'black')

for i in range (num_lines):
    angle = 2 * np.pi *i /num_lines
    x_line = [0 , np.cos(angle)]
    y_line  = [0 , np.sin(angle)] 
    ax.plot(x_line , y_line , color = 'blue', linewidth = 0.8 )


ax.axis('off')
plt.show()    
```

Nested if else 


```python
if a_number % 2 == 0:
    print("{} is even".format(a_number))
    if a_number % 3 == 0:
        print("{} is also divisible by 3".format(a_number))
    else:
        print("{} is not divisible by 3".format(a_number))
else:
    print("{} is odd".format(a_number))
    if a_number % 5 == 0:
        print("{} is also divisible by 5".format(a_number))
    else:
        print("{} is not divisible by 5".format(a_number))
```


While loop


```python
line = '*'
max_length = 10

while len(line) < max_length:
    print(line)
    line += "*"
    
while len(line) > 0:
    print(line)
    line = line[:-1]
```

```python
def crop_yeild(region , weights):
    result = 0
    for x , w in zip( region , weights):
        result += x* w 
    return result    
```
